```python
Table Of Content
Class-ATM
class ATM:
    def __init__(self):
        self.account = {'1234':{'saving':5000000, 'checking':980000}}
        self.__pin = None
        self.enter_pin()
    
    def get_pin(self):
        return self.__pin
    
    def set_pin(self, pin):
        self.__pin = pin
    
    def enter_pin(self):
        while True:
            pin1, pin2 = input("Enter pin: "), input("Verify pin: ")
            if (pin1 == pin2) and (pin1 in self.account):
                self.set_pin(pin1)
                self.choice()
            print("Enter pin again!")  # IN CASE OF ENTERING THE WRONG PIN!
    
    def choice(self):
        while True: # CHOOSE THE VALID CHOICE!
            print("----------------Welcome-----------------")
            print("Select 1 : To show your account balance")
            print("Select 2 : To deposit")
            print("Select 3 : To withdraw")
            print("Select 4 : To transfer")
            ch = input("Enter choice: ")
            if ch == '1':
                self.ch1()
            elif ch == '2':
                self.ch2()
            elif ch == '3':
                self.ch3()
            elif ch == '4':
                self.ch4()
            else:
                print("Invalid choice!") # IN CASE OF ENTERING THE WRONG CHOICE!
                self.enter_pin()
    
    def ch1(self):
        print("Your Account Balance is",self.account[self.get_pin()]['saving'])
    
    def ch2(self):
        deposit = int(input("Enter deposit money: "))
        self.account[self.get_pin()]['saving'] += deposit
        print("Your Money is Succesfully credited with {}  and your current balance is {}".format(deposit,self.account[self.get_pin()]['saving']))
    
    def ch3(self):
        withdraw = int(input("Enter withdraw money: ")) # WITHRAWING THE MONEY IF LESS THAN OR EQUAL TO AVAILABLE BALANCE!
        if self.account[self.get_pin()]['saving'] >= withdraw:
            self.account[self.get_pin()]['saving'] -= withdraw
            print("Your Savings account is debited with {} and your current account balance is {}.".format(withdraw,self.account[self.get_pin()]['saving']))
        else:
            print("Sorry You Dont have enough balance!")
    
    def ch4(self):
        transfer = int(input("Enter transfer amount: ")) # CREDIT/DEBIT THE AMOUNT INTO THE ACCOUNT!
        if self.account[self.get_pin()]['checking'] >= transfer:
            self.account[self.get_pin()]['checking'] += transfer
            self.account[self.get_pin()]['saving'] -= transfer
            print("Your Saving Account Balance is debited with {} and Your current account  balance is {}.".format(transfer,self.account[self.get_pin()]['saving']))
            print("Your account is Credited with {} and Your updated Checking balance is {} .".format(transfer,self.account[self.get_pin()]['checking']))
        else:
            print("Not enough balance!")
    
            

if __name__ == '__main__':
    atm = ATM() # CREATING THE OBJECT OF ATM CLASS!
```
